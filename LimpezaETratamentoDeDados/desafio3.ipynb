```python
with open('cursos_cadastrados.json') as f:
    dados_brutos = json.load(f)
dados_brutos
```

```python
pd.json_normalize(dados_brutos)
```

```python
df = pd.json_normalize(dados_brutos)
```

```python
df
```

```python
df.dropna(axis=0, subset='preco', inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
df
```

```python
df[df['concluintes'] == ""]
```

```python
df.drop(index=[2,3], inplace=True) # df.replace('', pd.NA, inplace=True) and then -> df.dropna()
df
```

```python
df.info()
```

```python
df['concluintes'] = df['concluintes'].astype(np.int64)
df['preco'] = df['preco'].astype(np.float64)
df['data_inicio'] = pd.to_datetime(df['data_inicio'], format='%Y-%m-%d')
df['data_conclusao'] = pd.to_datetime(df['data_conclusao'], format='%Y-%m-%d')
df
```
