# Lendo e Escrevendo Arquivos JSON

```python
dados_pacientes = pd.read_json('./pacientes.json')
dados_pacientes
```

```python
dados_pacientes2 = pd.read_json('https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes_2.json')
dados_pacientes2
```

```python
pd.json_normalize(dados_pacientes2)
```

```python
dados_pacientes_normalizados = pd.json_normalize(dados_pacientes2['Pacientes']) # Transforma uma coluna com valores json num DataSet
dados_pacientes_normalizados
```

```python
dados_pacientes_normalizados2 = pd.json_normalize(dados_pacientes2, record_path=['Pacientes']) # Mesmo efeito do cÃ³digo acima
dados_pacientes_normalizados2
```

```python
dados_pacientes_normalizados3 = pd.json_normalize(dados_pacientes2, record_path=['Pacientes'], meta=['Pesquisa', 'Ano']) # Mesmo efeito do cÃ³digo acima mas incluindo valores do nvl superior a 'Pacientes'.
dados_pacientes_normalizados3
```

A funÃ§Ã£o json_normalize() aceita apenas um dicionÃ¡rio ou uma lista de dicionÃ¡rios. Por isso, no vÃ­deo foi usada a estratÃ©gia de usar o cÃ³digo: pd.json_normalize(dados_pacientes_2['Pacientes']). 

PorÃ©m, se tentarmos usar parÃ¢metros da funÃ§Ã£o json_normalize em um arquivo JSON podem surgir erros. 

Para contornar isso, precisamos importar o mÃ³dulo json e ler os arquivos conforme o cÃ³digo abaixo:

```python
import json
```

```python
with open('pacientes_2.json', 'r') as f:
    dados_pacientes = json.loads(f.read())

dados_pacientes_normalizados4 = pd.json_normalize(dados_pacientes, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
dados_pacientes_normalizados4
```

```python
dados_pacientes_normalizados4.to_json('pacientes_normalizados.json')
```

```python
dados_pacientes_teste = pd.read_json('pacientes_normalizados.json')
dados_pacientes_teste
```

# Banco de Dados

```python
import sqlalchemy
```

```python
sqlalchemy.__version__
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
clientes_banco = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
clientes_banco.head()
```

```python
clientes_banco.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = "select * from clientes where [Categoria_de_renda] = 'Empregado'"
```

```python
empregados = pd.read_sql(query, engine)
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
query2 = "select * from clientes"
pd.read_sql(query2, engine)
```

```python
query3 = "delete from clientes where [ID_Cliente] = 5008804"
with engine.connect() as conn:
    conn.execute(query3)
```

```python
query4 = "update clientes set [Grau_escolaridade] = 'Ensino superior' where [ID_Cliente] = 5008808"
with engine.connect() as conn:
    conn.execute(query4)
```
